> To determine if a SAM file is sorted, we can check the header lines in the SAM file.
The header should contain a line that indicates the sorting order of the alignments.
###### Format: @HD     VN:1.6  SO:coordinate(for genomic postion)\ SO:queryname(read name) SO:unsorted no SO field
